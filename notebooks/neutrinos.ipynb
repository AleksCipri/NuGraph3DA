```python
import h5py
import numpy as np

def compare_hdf5_files(file1_path, file2_path, check_values=False):
    with h5py.File(file1_path, 'r') as f1, h5py.File(file2_path, 'r') as f2:
        def compare_groups(g1, g2, path='/'):
            keys1 = set(g1.keys())
            keys2 = set(g2.keys())

            # Compare keys (datasets or groups)
            if keys1 != keys2:
                print(f"Different keys at {path}")
                print(f"File1 has: {keys1 - keys2}")
                print(f"File2 has: {keys2 - keys1}")
                return False

            all_match = True

            for key in keys1:
                item1 = g1[key]
                item2 = g2[key]
                subpath = f"{path}{key}/"

                # Check if same type (Group or Dataset)
                if type(item1) != type(item2):
                    print(f"Type mismatch at {subpath}")
                    all_match = False
                    continue

                if isinstance(item1, h5py.Dataset):
                    # Check shape and dtype
                    if item1.shape != item2.shape:
                        print(f"Shape mismatch at {subpath}: {item1.shape} vs {item2.shape}")
                        all_match = False
                    elif item1.dtype != item2.dtype:
                        print(f"Dtype mismatch at {subpath}: {item1.dtype} vs {item2.dtype}")
                        all_match = False
                    elif check_values:
                        if not np.allclose(item1[()], item2[()], equal_nan=True):
                            print(f"Value mismatch at {subpath}")
                            all_match = False
                else:
                    # Recursively compare sub-groups
                    if not compare_groups(item1, item2, subpath):
                        all_match = False

            return all_match

        print(f"Comparing:\n  {file1_path}\n  {file2_path}")
        identical = compare_groups(f1, f2)
        if identical:
            print("Files are structurally identical.")
        else:
            print("Files differ.")
```

## Testing simple example

```python
import pytorch_lightning as pl
from torch.utils.data import DataLoader, Dataset
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F

# Dummy dataset
class DummyDataset(Dataset):
    def __init__(self, size=100):
        self.data = torch.randn(size, 10)  # 10 features per data point
        self.labels = torch.randint(0, 2, (size,))  # Binary labels

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx], self.labels[idx]  # Ensure returning as a tuple (data, label)

# Lightning DataModule with two identical dataloaders
class H2DataModule(pl.LightningDataModule):
    def __init__(self, dataset, batch_size=32, num_workers=4):
        super().__init__()
        self.dataset = dataset
        self.batch_size = batch_size
        self.num_workers = num_workers

    def train_dataloader(self) -> list[DataLoader]:
        # Create two data loaders, which load the same dataset
        loader_1 = DataLoader(self.dataset, batch_size=self.batch_size, num_workers=self.num_workers)
        loader_2 = DataLoader(self.dataset, batch_size=self.batch_size, num_workers=self.num_workers)
        # Return both loaders as a list
        return [loader_1, loader_2]

# Simple Encoder and Decoder network
class Encoder(nn.Module):
    def __init__(self):
        super(Encoder, self).__init__()
        self.fc = nn.Linear(10, 5)  # Encode to a 5-dim embedding

    def forward(self, x):
        return self.fc(x)

class Decoder(nn.Module):
    def __init__(self):
        super(Decoder, self).__init__()
        self.fc = nn.Linear(5, 2)  # Decode to output size 2 for cross-entropy

    def forward(self, x):
        return self.fc(x)

# Maximum Mean Discrepancy (MMD) loss
def mmd_loss(x1, x2):
    # Simple implementation of MMD using a Gaussian kernel
    def gaussian_kernel(x, y, sigma=1.0):
        x = x.unsqueeze(1)
        y = y.unsqueeze(0)
        return torch.exp(-((x - y).pow(2)).sum(2) / (2 * sigma ** 2))

    k_xx = gaussian_kernel(x1, x1).mean()
    k_yy = gaussian_kernel(x2, x2).mean()
    k_xy = gaussian_kernel(x1, x2).mean()
    return k_xx + k_yy - 2 * k_xy


```

```python
class MyLightningModule(pl.LightningModule):
    def __init__(self):
        super().__init__()
        self.encoder = Encoder()
        self.decoder = Decoder()
        self.encoded_batches = []

    def forward(self, x):
        return self.encoder(x)

    def training_step(self, batch, batch_idx):
        # Unpack the batch (should be a list of tensors)
        if isinstance(batch, list) and len(batch) == 2:
            data1, target1 = batch[0]  # First DataLoader output
            data2, target2 = batch[1]  # Second DataLoader output
        else:
            raise ValueError("Expected batch to be a list containing two batches.")

        # Ensure both data batches are tensors
        if not isinstance(data1, torch.Tensor) or not isinstance(data2, torch.Tensor):
            raise TypeError("Input data must be tensors")

        # Debug: Print the shape and type of the input data
        print(f"Data1 shape: {data1.shape}, Type: {type(data1)}")
        print(f"Data2 shape: {data2.shape}, Type: {type(data2)}")
        print(f"Target1 shape: {target1.shape}, Type: {type(target1)}")
        print(f"Target2 shape: {target2.shape}, Type: {type(target2)}")

        # Forward through the encoder
        encoded1 = self.encoder(data1)
        encoded2 = self.encoder(data2)

        # Debug: Print the shape of encoded outputs
        print(f"Encoded1 shape: {encoded1.shape}, Encoded2 shape: {encoded2.shape}")

        # Calculate Cross-Entropy Loss
        output = self.decoder(encoded1)
        ce_loss = F.cross_entropy(output, target1)
        self.log('train_ce_loss', ce_loss)

        # Calculate MMD Loss
        mmd_loss_val = torch.tensor(0.0, device=data1.device)  # Initialize MMD loss
        if len(self.encoded_batches) > 0:  # Ensure we have an encoded batch from the first dataloader
            with torch.no_grad():  # Ensure no gradient tracking for the previous batch
                mmd_loss_val = mmd_loss(self.encoded_batches[-1], encoded2)
            self.log('train_mmd_loss', mmd_loss_val)

        # Debug: Print the calculated loss values
        print(f"CE Loss: {ce_loss.item()}, MMD Loss: {mmd_loss_val.item()}")

        # Update the stored encoded batch
        self.encoded_batches.append(encoded1)

        # Calculate total loss
        total_loss = ce_loss + 0.1 * mmd_loss_val  # Scale MMD loss if needed
        self.log('train_total_loss', total_loss)

        # Return total loss for optimization
        return total_loss

    def configure_optimizers(self):
        return optim.Adam(self.parameters(), lr=1e-3)
```

```python
# Example usage
if __name__ == "__main__":
    dataset = DummyDataset(size=200)
    h2data_module = H2DataModule(dataset)
    model = MyLightningModule()

    # Trainer to run the training loop
    trainer = pl.Trainer(max_epochs=1, limit_train_batches=2)  # Automatically handles multiple dataloaders
    trainer.fit(model, h2data_module)
```

## Simple example wiht dummy graph data

```python
import torch
import torch.nn.functional as F
from torch import optim
import pytorch_lightning as pl
from torch_geometric.data import Data, Dataset, DataLoader
from torch_geometric.nn import GCNConv

# Dummy graph dataset using PyTorch Geometric
class GraphDummyDataset(Dataset):
    def __init__(self, num_samples, num_nodes, num_node_features):
        super(GraphDummyDataset, self).__init__()
        self.num_samples = num_samples
        self.num_nodes = num_nodes
        self.num_node_features = num_node_features

    def len(self):
        return self.num_samples

    def get(self, idx):
        # Create a random graph with random node features and random edge indices
        x = torch.randn(self.num_nodes, self.num_node_features)  # Node feature matrix
        edge_index = torch.randint(0, self.num_nodes, (2, self.num_nodes * 2), dtype=torch.long)  # Random edges
        y = torch.randint(0, 2, (self.num_nodes,))  # Random binary labels for nodes
        
        # Return a Data object which contains the graph info
        return Data(x=x, edge_index=edge_index, y=y)

# Graph Encoder using GCN
class GraphEncoder(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels):
        super(GraphEncoder, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        return x

# Core Network with single message-passing layer
class CoreNet(torch.nn.Module):
    def __init__(self, hidden_channels):
        super(CoreNet, self).__init__()
        self.conv = GCNConv(hidden_channels, hidden_channels)

    def forward(self, x, edge_index):
        x = self.conv(x, edge_index)
        x = F.relu(x)
        return x

# Graph Decoder (For simplicity, we'll keep it linear)
class GraphDecoder(torch.nn.Module):
    def __init__(self, in_channels, out_channels):
        super(GraphDecoder, self).__init__()
        self.fc = torch.nn.Linear(in_channels, out_channels)

    def forward(self, x):
        return self.fc(x)

# Modified Lightning Module to handle graph data
class MyLightningModule(pl.LightningModule):
    def __init__(self, num_iters=3):
        super().__init__()
        self.encoder = GraphEncoder(in_channels=10, hidden_channels=16)  # Assume 10 node features
        self.core_net = CoreNet(hidden_channels=16)  # Core network with single GCN layer
        self.decoder = GraphDecoder(in_channels=16, out_channels=2)   # Binary classification for nodes
        self.num_iters = num_iters  # Number of message-passing iterations
        self.encoded_batches = []

    def forward(self, data):
        # Forward through the encoder
        x = self.encoder(data)

        # Iterative message-passing using core network
        for _ in range(self.num_iters):
            x = self.core_net(x, data.edge_index)
            
        return x

    def training_step(self, batch, batch_idx):
        # Unpack the batch (should be a list of two graph data objects)
        if isinstance(batch, list) and len(batch) == 2:
            data1, data2 = batch[0], batch[1]  # First and second graph data samples
        else:
            raise ValueError("Expected batch to be a list containing two DataLoader outputs.")

        # Forward through encoder and iterative core network
        encoded1 = self.encoder(data1)
        for _ in range(self.num_iters):
            encoded1 = self.core_net(encoded1, data1.edge_index)
        
        encoded2 = self.encoder(data2)
        for _ in range(self.num_iters):
            encoded2 = self.core_net(encoded2, data2.edge_index)

        # Forward through the decoder
        output = self.decoder(encoded1)

        # Calculate Cross-Entropy Loss (binary node classification)
        ce_loss = F.cross_entropy(output, data1.y)
        self.log('train_ce_loss', ce_loss)

        # Calculate MMD Loss
        mmd_loss_val = torch.tensor(0.0, device=data1.x.device)  # Initialize MMD loss
        if len(self.encoded_batches) > 0:  # Ensure we have an encoded batch from the first dataloader
            with torch.no_grad():  # Ensure no gradient tracking for the previous batch
                mmd_loss_val = mmd_loss(self.encoded_batches[-1], encoded2)
            self.log('train_mmd_loss', mmd_loss_val)

        # Update the stored encoded batch
        self.encoded_batches.append(encoded1)

        # Calculate total loss
        total_loss = ce_loss + 0.1 * mmd_loss_val  # Scale MMD loss if needed
        self.log('train_total_loss', total_loss)

        # Return total loss for optimization
        return total_loss

    def configure_optimizers(self):
        return optim.Adam(self.parameters(), lr=1e-3)

# Dummy MMD loss function (assumed to be correctly defined elsewhere)
def mmd_loss(encoded1, encoded2):
    # Dummy MMD loss function for illustration
    return torch.mean((encoded1 - encoded2) ** 2)

# Example usage
if __name__ == "__main__":
    dataset = GraphDummyDataset(num_samples=1000, num_nodes=32, num_node_features=10)
    train_loader = DataLoader(dataset, batch_size=32, shuffle=True)

    model = MyLightningModule(num_iters=3)  # Set number of message-passing iterations
    trainer = pl.Trainer(max_epochs=10)
    trainer.fit(model, train_loader)
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
from torch_geometric.data import Data, Batch

# Dummy Dataset A
class DatasetA(Dataset):
    def __init__(self, size=100):
        self.size = size

    def __len__(self):
        return self.size

    def __getitem__(self, idx):
        # Create a dummy Data object with random features
        return Data(
            x=torch.rand(10, 3),  # 10 nodes, 3 features each
            edge_index=torch.randint(0, 10, (2, 20)),  # 20 edges
            y=torch.tensor(idx)  # Dummy target
        )


# Dummy Dataset B
class DatasetB(Dataset):
    def __init__(self, size=100):
        self.size = size

    def __len__(self):
        return self.size

    def __getitem__(self, idx):
        # Create another dummy Data object with random features
        return Data(
            x=torch.rand(15, 2),  # 15 nodes, 2 features each
            edge_index=torch.randint(0, 15, (2, 30)),  # 30 edges
            y=torch.tensor(idx + 100)  # Dummy target
        )


# CombinedDataset for pairing elements from DatasetA and DatasetB
class CombinedDataset(Dataset):
    def __init__(self, datasetA, datasetB):
        self.datasetA = datasetA
        self.datasetB = datasetB

    def __len__(self):
        # Use the minimum length of the two datasets
        return min(len(self.datasetA), len(self.datasetB))

    def __getitem__(self, idx):
        dataA = self.datasetA[idx]
        dataB = self.datasetB[idx]
        return dataA, dataB  # Return both data objects


# Collate function to batch the combined dataset
def collate_fn(batch):
    dataA_list, dataB_list = zip(*batch)  # Unzip dataA and dataB
    batchA = Batch.from_data_list(dataA_list)  # Batch data from datasetA
    batchB = Batch.from_data_list(dataB_list)  # Batch data from datasetB
    return batchA, batchB


# Initialize the datasets
datasetA = DatasetA(size=100)
datasetB = DatasetB(size=100)

# Combine the datasets
combined_dataset = CombinedDataset(datasetA, datasetB)

# Create a DataLoader with the collate function
loader = DataLoader(combined_dataset, batch_size=32, shuffle=True, collate_fn=collate_fn)

# Example usage
for batchA, batchB in loader:
    print("Batch A:")
    print(batchA)
    print("Batch B:")
    print(batchB)
    break  # Exit after one batch to keep output short
```

```python
import torch
import tqdm
import torchmetrics as tm
import plotly.graph_objects as go

class Score(tm.Metric):
    def __init__(self,
                 num_classes: int,
                 bins: int = 20,
                 range: tuple[float] = (0,1),
                 ignore_index: int = None):
        super().__init__()

        self.bins = bins
        self.range = range
        self.ignore_index = ignore_index
        
        self.add_state('true', default=torch.zeros(num_classes, bins), dist_reduce_fx = 'sum')
        self.add_state('false', default=torch.zeros(num_classes, bins), dist_reduce_fx = 'sum')

    def update(self, preds: torch.Tensor, target: torch.Tensor):

        # check that number of classes is correct
        num_classes = preds.size(1)
        assert num_classes == self.true.size(0)

        # loop over labels
        filter = (target != self.ignore_index)
        for label in range(num_classes):
            mask = filter & (target == label)
            hist, bin_edges = preds[mask, label].histogram(bins=self.bins,
                                                           range=self.range)
            self.true[label] += hist
            mask = filter & (target != label)
            hist, bin_edges = preds[mask, label].histogram(bins=self.bins,
                                                           range=self.range)
            self.false[label] += hist
        
    def compute(self):
        true = self.true / self.true.sum(dim=1)[:,None]
        false = self.false / self.false.sum(dim=1)[:,None]
        return true, false

style = {
    'layout_width': 800,
    'layout_height': 450,
    'layout_margin_b': 20,
    'layout_margin_t': 20,
    'layout_margin_r': 20,
    'layout_margin_l': 20,
    'layout_xaxis_title_font_size': 24,
    'layout_xaxis_tickfont_size': 20,
    'layout_yaxis_title_font_size': 24,
    'layout_yaxis_tickfont_size': 20,
    'layout_legend_font_size': 20,
}
```

```python
num_classes = len(nudata.semantic_classes)

score = Score(num_classes=num_classes, ignore_index=-1)
roc_filter = tm.classification.BinaryROC(
    thresholds=1000,
)
roc_semantic = tm.classification.MulticlassROC(
    num_classes=num_classes,
    thresholds=1000,
    ignore_index=-1,
)

batches = trainer.predict(model, nudata.test_dataloader())

#source
for batch in tqdm.tqdm(batches[0]):
    for p in nudata.planes:
        score.update(batch[p].x_semantic, batch[p].y_semantic)
        roc_filter.update(batch[p].x_filter, batch[p].y_semantic!=-1)
        roc_semantic.update(batch[p].x_semantic, batch[p].y_semantic)

true, false = score.compute()
fpr_filter, tpr_filter, thresholds = roc_filter.compute()
fpr_semantic, tpr_semantic, thresholds = roc_semantic.compute()


# #target
# for batch in tqdm.tqdm(batches[1]):
#     for p in nudata.planes:
#         score.update(batch[p].x_semantic, batch[p].y_semantic)
#         roc_filter.update(batch[p].x_filter, batch[p].y_semantic!=-1)
#         roc_semantic.update(batch[p].x_semantic, batch[p].y_semantic)

# true, false = score.compute()
# fpr_filter, tpr_filter, thresholds = roc_filter.compute()
# fpr_semantic, tpr_semantic, thresholds = roc_semantic.compute()
```

## Testing isomaps in tensorflow

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.manifold import Isomap
from sklearn.preprocessing import LabelEncoder
from io import BytesIO
import torch
from PIL import Image

class IsomapCombinedPlot:
    def __init__(self, n_neighbors=10, n_components=2, writer=None):
        """
        Initializes the IsomapCombinedPlot class.

        Parameters:
            n_neighbors: int, optional (default=5)
                Number of neighbors for Isomap.
            n_components: int, optional (default=2)
                Number of dimensions for Isomap projection.
            writer: torch.utils.tensorboard.SummaryWriter, optional
                TensorBoard SummaryWriter for logging images.
        """
        self.n_neighbors = n_neighbors
        self.n_components = n_components
        self.writer = writer  # TensorBoard SummaryWriter instance

    def plot_isomap_combined_concatenated(self, dataset1, labels1, dataset2, labels2, epoch=None):
        """
        Plots Isomap embeddings of two datasets with different point shapes for each dataset.
        Isomap fitting is performed on the concatenated datasets to preserve relative distances.

        Parameters:
            dataset1: array-like, shape (n_samples1, n_features)
                First dataset.
            labels1: array-like, shape (n_samples1,)
                Class labels for the first dataset.
            dataset2: array-like, shape (n_samples2,)
                Second dataset.
            labels2: array-like, shape (n_samples2,)
                Class labels for the second dataset.
            epoch: int, optional
                Current training epoch (used for TensorBoard logging).
        """
        # Convert PyTorch tensors to NumPy arrays for Isomap
        
        # If there is time I can try implementing GPU version of this (matplotlib is CPU-based) 
        #but I need to install some packages:
        #conda install -c rapidsai -c nvidia -c conda-forge \
        #cuml=21.12 python=3.8 cudatoolkit=11.5
        
        dataset1_np = dataset1.detach().cpu().numpy()
        labels1_np = labels1.detach().cpu().numpy()
        dataset2_np = dataset2.detach().cpu().numpy()
        labels2_np = labels2.detach().cpu().numpy()
        
        # Concatenate datasets
        combined_data = np.concatenate([dataset1_np, dataset2_np], axis=0)
        combined_labels = np.concatenate([labels1_np, labels2_np])

        # Fit Isomap on the combined data
        isomap = Isomap(n_neighbors=self.n_neighbors, n_components=self.n_components)
        combined_embedding = isomap.fit_transform(combined_data)

        # Split the embeddings back into separate datasets
        embedding1 = combined_embedding[:len(dataset1_np)]
        embedding2 = combined_embedding[len(dataset1_np):]

        # Unique classes across datasets
        unique_classes = np.unique(combined_labels)

        # Generate a colormap for all unique classes
        colors = plt.cm.inferno(np.linspace(0, 1, len(unique_classes)))
        class_color_map = {cls: colors[idx] for idx, cls in enumerate(unique_classes)}

        # Create the plot
        plt.figure(figsize=(12, 10))

        # Plot dataset1 with circles
        for cls in np.unique(labels1_np):
            mask = labels1_np == cls
            plt.scatter(
                embedding1[mask, 0], embedding1[mask, 1],
                label=f"Source - Class {cls}",
                color=class_color_map[cls],
                marker='o', s=100, edgecolor='k', alpha=0.8
            )

        # Plot dataset2 with triangles
        for cls in np.unique(labels2_np):
            mask = labels2_np == cls
            plt.scatter(
                embedding2[mask, 0], embedding2[mask, 1],
                label=f"Target - Class {cls}",
                color=class_color_map[cls],
                marker='^', s=150, edgecolor='k', alpha=0.8
            )

        # Add legend and labels
        plt.legend(fontsize=12, loc='upper right', bbox_to_anchor=(1, 1))
        plt.title("Isomap Embeddings of Source and Target Datasets", fontsize=16)

        # Save plot to TensorBoard
        if self.writer:
            buf = BytesIO()
            plt.savefig(buf, format='png')
            buf.seek(0)
            img = Image.open(buf)
            img_tensor = torch.tensor(np.array(img)).permute(2, 0, 1)  # Convert to CHW format
            self.writer.add_image('Isomap Plot', img_tensor, global_step=epoch)

        plt.close()
```

```python
from torch.utils.tensorboard import SummaryWriter

# Example datasets as PyTorch tensors
data1 = torch.randn(100, 10, device='cuda')
labels1 = torch.randint(0, 3, (100,), device='cuda')
data2 = torch.randn(100, 10, device='cuda')
labels2 = torch.randint(0, 5, (100,), device='cuda')

# Initialize TensorBoard writer
writer = SummaryWriter(log_dir='runs/isomap_gpu_plot')

# Initialize the plotter
isomap_plotter = IsomapCombinedPlot(n_neighbors=5, n_components=2, writer=writer)

# Simulate training process
for epoch in range(10):
    isomap_plotter.plot_isomap_combined_concatenated(data1, labels1, data2, labels2, epoch=epoch)

# Close the writer
writer.close()
```
